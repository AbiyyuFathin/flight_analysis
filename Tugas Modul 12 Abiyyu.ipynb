```python
#Membaca berkas .csv menggunakan pandas, kemudian ditampilkan 6 baris pertama dari berkas tersebut.

flights = pd.read_csv("C:/Users/abiyy/Downloads/Material_Modul_12_DTS_AI/flights.csv")
flights.head(6)
```

```python
#2. Sebutkan nama-nama carrier-nya secara unik!

flights['carrier'].unique()

#Nama carrier yang unik dari data "flight.csv" adalah UA', 'AA', 'B6', 'DL', 'EV', 'MQ', 'US', 'WN', 'VX', 'FL', 'AS','9E', 
#'F9', 'HA', 'YV', 'OO
#Saya menggunakan method unique karena saya ingin menampilkan nama kolom dengan tidak berulang.
```

```python
#3. Sebutkan nama-nama carrier dan tailnumnya!

flights.loc[:,['carrier', 'tailnum']]

#Nama carrier dan tailnum nya sesuai seperti yang ditampilkan pada Data Frame di bawah ini.
#Saya menggunakan method loc untuk mengambil hanya kolom carrier dan tailnum. 
```

```python
# 4. Berapa delay paling lama yang pernah terjadi?

flights['arr_delay'].max()
```

```python
flights['dep_delay'].max()

#Delay paling lama terjadi pada dep_delay, yaitu 1301.
#Saya menggunakan method max pada kedua delay, lalu mengambil hasil terbesar dari kedua jawaban tersebut.
#Hal ini guna menghemat komputasi daripada menggunakan sort.
```

```python
#5. Pada bulan apa terdapat penerbangan terbanyak?

No5 = flights.groupby(['month'])['month'].count()
No5
```

```python
No5.max()

#Maka, penerbangan terbanyak terjadi pada bulan 7 atau Juli.
#Saya menggunakan grouoby untuk menyeleksi bulan dan menghitung masing-masing bulan menggunakan method count. Lalu, saya cari
#nilai maksimumnya.
```

```python
#6. Berapa rata-rata delay (departure delay) pesawat dari "JFK"?

flights.groupby(flights['origin'] == 'JFK')[['dep_delay']].mean()

#Adapun rata-rata departure delay pesawat dari "JFK" adalah 12.112159. 
#Saya menggunakan groupby untuk menyeleksi kolom origin saja, kemudian menggunakan filter untuk mengambil nilai origin JFK saja.
#Setelah itu, saya hitung dep_delay nya menjadi nilai rata-ratanya.
#Saya menggunakan nilai True karena nilai tersebut yang mewakili hasil rerata sebenarnya. Adapn untuk hasil False, adalah hasil 
#dari nilai selian "JFK".
```

```python
#7. Pesawat apa (sebutkan tailnumnya) yang paling sering mengalami keterlambatan datang (arr_delay)?

No7 = flights.groupby(['tailnum'])[['dep_delay']].count()
No7.sort_values(by =['dep_delay'], ascending = False).head(3)

#Adapun pesawat N725MQ merupakan pesawat yang paling sering mengalami keterlambatan datang.
#Saya menggunakan groupby untuk menyeleksi kolom tailnum saja. Lalu, saya menghitung berapa banyak dep_delay menggunakan method
#count. Setelah itu, saya urutkan datanya menggunakan sort_values dari nilai dep_delay secara descending. Lalu, saya tampilkan
#3 nilai awalnya.
```

```python
#8. Pesawat apa (sebutkan tailnumnya) yang paling lama rerata durasi keterlambatan datang (arr_delay)-nya?

No8 = flights.groupby(['tailnum'])[['arr_delay']].mean()
No8.sort_values(by =['arr_delay'], ascending = False).head(3)

#Adapun pesawat N844MH merupakan pesawat yang paling lama rerata durasi keterlambatan datangnya.
#Saya menggunakan groupby untuk menyeleksi kolom tailnum saja. Lalu, saya menghitung berapa rerata arr_delay menggunakan method
#mean. Setelah itu, saya urutkan datanya menggunakan sort_values dari nilai arr_delay secara descending. Lalu, saya tampilkan
#3 nilai awalnya.
```
